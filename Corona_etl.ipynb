```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
csv_file = "sources/time_series_covid19_confirmed_global.csv"
confirmed_cases_df = pd.read_csv(csv_file)
confirmed_cases_df.head()
```

```python
csv_file = "sources/time_series_covid19_deaths_global.csv"
deaths_df = pd.read_csv(csv_file)
deaths_df.head()
```

```python
csv_file = "sources/time_series_covid19_recovered_global.csv"
recovered_cases_df = pd.read_csv(csv_file)
recovered_cases_df.head()
```

### Transform DataFrame

```python
#List column header to check if there are different entries in each file
confirmed_col = confirmed_cases_df.columns
deaths_col= deaths_df.columns
recovered_col = recovered_cases_df.columns

difference_test1 = confirmed_col.difference(deaths_col)
difference_test1
```

```python
#Remove province category since it has limited and different entries in each file
confirmed_cases_df = confirmed_cases_df.drop(columns=["Province/State"])
```

```python
deaths_df.columns.difference(confirmed_cases_df.columns)
```

```python
recovered_cases_df.columns.difference(confirmed_cases_df.columns)
```

```python
deaths_df = deaths_df.drop(columns=["Province/State"])
recovered_cases_df = recovered_cases_df.drop(columns=["Province/State"])
```

```python
confirmed_cases_df["Country/Region"].value_counts()
```

```python
deaths_df["Country/Region"].value_counts()
```

```python
recovered_cases_df["Country/Region"].value_counts()
```

```python
recovered_cases_df = recovered_cases_df.groupby(["Country/Region"])

recovered_cases_df.get_group("Canada")
```

```python
#Noted difference between recovered dataframe and confirmed/deaths is in the umber of entries for Canada
confirmed_cases_df =confirmed_cases_df.groupby(["Country/Region"])

confirmed_cases_df.get_group("Canada")
```

```python
#created dataframes based on the sum of cases by country for each day across the files /double check with number of rows
confirmed_cases_sum_df = confirmed_cases_df.sum()
confirmed_cases_sum_df
```

```python
#Repeat the above for deaths_df
deaths_df = deaths_df.groupby(["Country/Region"])
deaths_sum_df = deaths_df.sum()
deaths_sum_df
```

```python
#Repeat the above for recovered_cases_df
recovered_sum_cases_df = recovered_cases_df.sum()
recovered_sum_cases_df
```

```python
confirmed_cases_sum_df = confirmed_cases_sum_df.sort_values(by=["Country/Region"])
deaths_sum_df = deaths_sum_df.sort_values(by=["Country/Region"])
recovered_sum_cases_df = recovered_sum_cases_df.sort_values(by=["Country/Region"])
```

```python
confirmed_cases_sum_df = confirmed_cases_sum_df.reset_index()
confirmed_cases_sum_df
```

```python
recovered_sum_cases_df = recovered_sum_cases_df.reset_index()
recovered_sum_cases_df 
```

```python
deaths_sum_df = deaths_sum_df.reset_index()
deaths_sum_df.index.name = 'ID'
deaths_sum_df
```

```python
recovered_sum_cases_df.index.name = 'ID'
recovered_sum_cases_df
```

```python
confirmed_cases_sum_df.index.name = 'ID'
confirmed_cases_sum_df
```
